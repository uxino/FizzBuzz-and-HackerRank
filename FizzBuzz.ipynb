# Fizz Buzz
# Fizz en az 5 en az 3 Buzz en azn 3FizzBuzz  

```python

def fizzBuzz(f,b,fB):
    fizz,buzz,fizzBuzz,x = 0,0,0,0
    while fizz<f or buzz<b or fizzBuzz<fB:
        if x%3==0 and x%15!=0:
            fizz+=1
            print(x,"Fizz")
        elif x%5==0 and x%15!=0:
            buzz+=1
            print(x,"Buzz")
        elif x%15==0 and x!=0:
            fizzBuzz+=1
            print(x,"FizzBuzz")
        else:
            print(x)
        x+=1
f,b,fB = [int(x) for x in input("fizz, buzz ve fizzbuzz değerlerini giriniz").split()]
fizzBuzz(f,b,fB)
    
        
```
